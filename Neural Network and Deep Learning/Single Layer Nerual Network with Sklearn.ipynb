```python
# reading data from csv files and converting to matrix 

import pandas as pd 
xtrain=pd.read_table('C:/Human_Activity_Recognition/dataset/train/X_train.txt',delim_whitespace=True,header=None)
xtest=pd.read_table('C:/Human_Activity_Recognition/dataset/test/X_test.txt',delim_whitespace=True,header=None)
ytrain=pd.read_table('C:/Human_Activity_Recognition/dataset/train/y_train.txt',header=None)
ytest=pd.read_table('C:/Human_Activity_Recognition/dataset/test/y_test.txt',header=None)
```

```python
# encoding labels 

from sklearn import preprocessing

encoder = preprocessing.LabelEncoder()

# encoding test labels 
encoder.fit(ytest)
testLabelE = encoder.transform(ytest)

# encoding train labels 
encoder.fit(ytrain)
trainLabelE = encoder.transform(ytrain)
```

### Try Sklearn's Nerual Network(Multi-layer perceptron)

```python
# applying supervised neural network using multi-layer perceptron 

import sklearn.neural_network as nn 

#activation : {‘identity’, ‘logistic’, ‘tanh’, ‘relu’}, default ‘relu’

#Activation function for the hidden layer.

# ‘identity’, no-op activation, useful to implement linear bottleneck, returns f(x) = x
# ‘logistic’, the logistic sigmoid function, returns f(x) = 1 / (1 + exp(-x)).
# ‘tanh’, the hyperbolic tan function, returns f(x) = tanh(x).
# ‘relu’, the rectified linear unit function, returns f(x) = max(0, x)

mlpSGD  =  nn.MLPClassifier(hidden_layer_sizes=(90,)  \
                        , max_iter=1000 , alpha=2.1#power of regularization  \
                        , solver='sgd' , verbose=10   \
                        , tol=1e-20 , random_state=1  \
                        , learning_rate_init=.001) 

mlpADAM =  nn.MLPClassifier(hidden_layer_sizes=(90,)  \
                        , max_iter=1000 , alpha=1.5  \
                        , solver='adam' , verbose=10  \
                        , tol=1e-20 , random_state=1  \
                        , learning_rate_init=.001) 

nnModelSGD  = mlpSGD .fit(xtrain , trainLabelE)
print('ADAM Solver Start!!!!!!!!!!!!!!!!')
nnModelADAM = mlpADAM.fit(xtrain , trainLabelE)
```

Note: The default solver ‘adam’ works pretty well on relatively large datasets (with thousands of training samples or more) in terms of both training time and validation score. For small datasets, however, ‘lbfgs’ can converge faster and perform better.

```python
# ploting nn convergence and testing score 

import matplotlib.pyplot as plt 
import numpy as np

X1 = np.linspace(1, nnModelSGD.n_iter_  , nnModelSGD.n_iter_ )
X2 = np.linspace(1, nnModelADAM.n_iter_ , nnModelADAM.n_iter_)

plt.plot(X1 , nnModelSGD.loss_curve_ , label = 'SGD Convergence' )
plt.plot(X2 , nnModelADAM.loss_curve_, label = 'ADAM Convergence')
plt.title('Error Convergence ')
plt.ylabel('Cost function')
plt.xlabel('Iterations')
plt.legend()
plt.show()
```

```python
# generating test scores for both classifiers 

print("Training set score for SDG : %f" % mlpSGD .score(xtrain, trainLabelE))
print("Training set score for ADAM: %f" % mlpADAM.score(xtrain, trainLabelE))
print("Test set score for SDG : %f"     % mlpSGD .score(xtest , testLabelE ))
print("Test set score for ADAM: %f"     % mlpADAM.score(xtest , testLabelE ))
```

```python
ypred = mlpADAM.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
confusion = confusion_matrix(testLabelE, ypred)
plt.figure(figsize=(10, 7))
sns.heatmap(confusion, annot=True)
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(testLabelE,ypred))
```
